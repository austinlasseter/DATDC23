## Bonus

Define a function that accepts a row of data and returns a predicted species. Then, use that function to make predictions for all existing rows of data, and check the accuracy of your predictions.

```python
def mymodel(petal_length):
    if petal_length <= 2.1:
        return 0
    elif petal_length >= 4.7:
        return 2
    else:
        return 1
```

```python
df['predictions'] = df['petal length'].apply(lambda x: mymodel(x))
```

```python
df[df['target'] == df['predictions']].shape[0] / float(df.shape[0])
```

```python
df['correct'] = df['target'] - df['predictions']
```
