## Apply Functions!

Three steps:

1. Define a function
2. Test that function on a single value
3. Use that function across a full column (or your money back guaranteed)

```python
# step one!!!!!!!!!!!!!!!
def pittiful(text):
    if "Brad Pitt" in text:
        return 1
    else:
        return 0
```

```python
# should be 1
movies[movies.title == 'Fight Club']
```

```python
# should be 1
movies[movies.title == 'Se7en']
```

```python
# should be 0
movies[movies.title == 'The Godfather']
```

```python
# Identify text with Brad Pitt in it
movies.actors_list[9]
```

```python
# test
pittiful(movies.actors_list[9])
```

```python
# Identify text without Brad Pitt in it
movies.actors_list[1]
```

```python
# Test text without Brad Pitt in it
pittiful(movies.actors_list[1])
```

```python
# WRONG - we are passing a series here
pittiful(movies.actors_list)
```

```python
# apply to a column
start_time = time.time()
movies['bradley'] = movies.actors_list.apply(pittiful)
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
# did it work?
movies.head()
```

```python
# get movies with just Brad pitt
movies[movies.bradley == 1]
```

```python
# define a dictionary
categories = {
    "Crime" : "Cri",
    "Action" : "Act",
    "Drama" : "John", 
    "Comedy": "Com",
    "Biography" : "Bio"    
}
```

Another example, stolen from [Chris Albon](https://chrisalbon.com/python/pandas_map_values_to_values.html)

```python
raw_data = {'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'], 
        'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'], 
        'age': [42, 52, 36, 24, 73], 
        'city': ['San Francisco', 'Baltimore', 'Miami', 'Douglas', 'Boston']}
df = pd.DataFrame(raw_data, columns = ['first_name', 'last_name', 'age', 'city'])
df
```

```python
city_to_state = { 'San Francisco' : 'California', 
                  'Baltimore' : 'Maryland', 
                  'Miami' : 'Florida', 
                  'Douglas' : 'Arizona', 
                  'Boston' : 'Massachusetts'}
```

```python
df['state'] = df['city'].map(city_to_state)
df
```
