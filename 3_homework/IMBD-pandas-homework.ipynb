```python
# read in 'imdb_1000.csv' and store it in a DataFrame named movies
df = pd.read_csv('../2_dataset/imdb_1000.csv')
```

```python
df.head()
```

```python
actors_split = []
for i in df.index.values:
    for actor in df.actors_list[i].strip("[]").split(", "):
        row = {}
        row['key'] = i
        row['actor'] = actor
        actors_split.append(row)
        
new_df = df.merge(pd.DataFrame(actors_split),left_index=True,right_on=['key']).drop(['actors_list','key'],axis=1)
```
